## Gaussian Distribution

```python
import matplotlib.mlab as mlab
import math
import scipy.stats

mu = 0
variance = 5
sigma = math.sqrt(variance)
x = np.linspace(mu - 5*sigma,mu+5*sigma,100)
plt.plot(x,scipy.stats.norm.pdf(x,mu,sigma))
plt.show()
```

## Gaussian Properties
### Multiplication

#### example
$$P(X|Z)=\frac{P(Z|X)P(X)}{P(Z)}$$
P(X|Z) is Posterior, P(X), P(Z) is prior, P(Z|X) is measurement  
Posterior를 계산하기 위해 P(Z|X)xP(X)을 계산하는 경우가 발생할 수 있음.
$$P(Z|X) -> N(\mu_{1},\sigma_{1})$$
$$P(X) -> N(\mu_{2},\sigma_{2})$$
일때,
    $$\mu_{new}=\frac{\sigma_{1}^2\mu_{2}+\sigma_{2}^2\mu_{1}}{\sigma_{1}^2+\sigma_{2}^2} $$
    $$\sigma_{new}=\sqrt{\frac{\sigma_{1}^2\sigma_{2}^2}{\sigma_{1}^2+\sigma_{2}^2}}$$
    
+alpha : Bayes filter

```python
## Multiplication 에제
import matplotlib.mlab as mlab
import math
mu1 = 0
variance1 = 2
sigma = math.sqrt(variance1)
x1 = np.linspace(mu1-3*sigma,mu1+3*sigma,100)
plt.plot(x1,scipy.stats.norm.pdf(x1,mu1,sigma),label='prior')

mu2 = 10
variance2 = 2
sigma = math.sqrt(variance2)
x2 = np.linspace(mu2-3*sigma,mu2+3*sigma,100)
plt.plot(x2,scipy.stats.norm.pdf(x2,mu2,sigma),label='measurement')

mu_new=(mu1*variance2+mu2*variance1)/(variance1+variance2)
print("New mean is at: ",mu_new)
var_new=(variance1*variance2)/(variance1+variance2)
print("New variance is: ",var_new)

sigma=math.sqrt(var_new)
x3 = np.linspace(mu_new - 3*sigma,mu_new+3*sigma,100)
plt.plot(x3,scipy.stats.norm.pdf(x3,mu_new,var_new),label='posterior')
plt.legend(loc='upper left')
plt.xlim(-10,20)
plt.show()
```

### Addition
두 Gaussian을 서로 더하는 경우,
$$ \mu_{new}=\mu_{1}+\mu_{2} $$
$$ \sigma^2_{new}=\sigma^2_{1}+\sigma^2_{2} $$

```python
import matplotlib.mlab as mlab
import math
mu1 = 5
variance1 = 1
sigma = math.sqrt(variance1)
x1 = np.linspace(mu1 - 3*sigma, mu1 + 3*sigma, 100)
plt.plot(x1,scipy.stats.norm.pdf(x1, mu1, sigma),label='prior')

mu2 = 10
variance2 = 1
sigma = math.sqrt(variance2)
x2 = np.linspace(mu2 - 3*sigma, mu2 + 3*sigma, 100)
plt.plot(x2,scipy.stats.norm.pdf(x2, mu2, sigma),"g-",label='measurement')


mu_new=mu1+mu2
print("New mean is at: ",mu_new)
var_new=(variance1+variance2)
print("New variance is: ",var_new)
sigma = math.sqrt(var_new)
x3 = np.linspace(mu_new - 3*sigma, mu_new + 3*sigma, 100)
plt.plot(x3,scipy.stats.norm.pdf(x3, mu_new, var_new),label="posterior")
plt.legend(loc='upper left')
plt.xlim(-10,20)
plt.show()
```

참고:einsum

- einsum은 더 공부할 필요가 있음.

```python
#Example from:
#https://scipython.com/blog/visualizing-the-bivariate-gaussian-distribution/
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
# X,Y 2차원 Random Variable에 대한 분포
# Our 2-dimensional distribution will be over variables X and Y

# X,Y 2차원 Random Variable 생성
N = 60
X = np.linspace(-3, 3, N)
Y = np.linspace(-3, 4, N)
X, Y = np.meshgrid(X, Y)

# Mean vector and covariance matrix
# 평균 벡터와 공분산 행렬 생성(아래에 정해준 값으로 생성)
mu = np.array([0., 1.])
Sigma = np.array([[ 1. , -0.5], [-0.5,  1.5]])


# Pack X and Y into a single 3-dimensional array
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y


def multivariate_gaussian(pos, mu, Sigma):
    """Return the multivariate Gaussian distribution on array pos.

    pos is an array constructed by packing the meshed arrays of variables
    x_1, x_2, x_3, ..., x_k into its _last_ dimension.

    """

    n = mu.shape[0]
    Sigma_det = np.linalg.det(Sigma)
    
    Sigma_inv = np.linalg.inv(Sigma)
    N = np.sqrt((2*np.pi)**n * Sigma_det)
    # This einsum call calculates (x-mu)T.Sigma-1.(x-mu) in a vectorized
    # way across all the input variables.
    fac = np.einsum('...k,kl,...l->...', pos-mu, Sigma_inv, pos-mu)

    return np.exp(-fac / 2) / N
    ## <-- Multivariate_gaussian 직접 구현한 사례
    
# The distribution on the variables X, Y packed into pos.
Z = multivariate_gaussian(pos, mu, Sigma)

# Create a surface plot and projected filled contour plot under it.
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, rstride=3, cstride=3, linewidth=1, antialiased=True,
                cmap=cm.viridis)

cset = ax.contourf(X, Y, Z, zdir='z', offset=-0.15, cmap=cm.viridis)

# Adjust the limits, ticks and view angle
ax.set_zlim(-0.15,0.2)
ax.set_zticks(np.linspace(0,0.2,5))
ax.view_init(27, -21)

plt.show()
```
